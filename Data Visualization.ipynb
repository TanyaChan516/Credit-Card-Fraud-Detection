### Module 2: Data Visualization
Explore and visualize the distributions of variables in the Fraud group and in the Normal group. 

The best way to check the correlation and connection between V1 to V28 with Class is by using a heatmap.

```python
corrmat = train_new.corr() 
fig = plt.figure(figsize = (16, 10)) 
sns.heatmap(corrmat, vmax = .8, square = True) 
plt.show() 
```

From the above graph, we will want to see the connections between V1 to V28 with Class, which obviously V14 and V17 highly negatively correlated with Fraud class.

```python
x = train_new['V14']
y = train_new['V17']
fig, ax = plt.subplots(figsize=(10, 10))
#applies the custom color map along with the color sequence
ax.scatter(x, y, alpha=0.70, c= train_new['Class'], cmap=cm.brg)

#adds a title and axes labels
ax.set_title('V14 VS V17 with Class')
ax.set_xlabel('V14')
ax.set_ylabel('V17')

#removing top and right borders
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
#adds major gridlines
ax.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.5)
plt.show()
```

From the above graph, we can conclude that the distributions of V14 and V17 in new Valid is very stable with a range -2.5 to 2.5 of both V14 and V17, while most data besides this range are Fraud. We can also say that either V14 or V17 is not within that small range, then it will be a Fraud most likely since they spread a lot.

```python
x = train_new['V2']
y = train_new['V11']
fig, ax = plt.subplots(figsize=(10, 10))
#applies the custom color map along with the color sequence
ax.scatter(x, y, alpha=0.70, c= train_new['Class'], cmap=cm.brg)

#adds a title and axes labels
ax.set_title('V2 VS V11 with Class')
ax.set_xlabel('V2')
ax.set_ylabel('V11')

#removing top and right borders
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
#adds major gridlines
ax.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.5)
plt.show()
```
